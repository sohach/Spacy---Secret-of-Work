### Rule-Based Matching Using spaCy


```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)

pattern1 = [{'TEXT':'Buddha'}]
pattern2 = [{'LOWER':'sacred'},{'LOWER':'person'}]
pattern3 = [{'TEXT':'India'}]

matcher.add('Secret of work',None,pattern1,pattern2,pattern3)

doc = nlp('Buddha was a working Jnni, Now in India a guest is a sacred person.')

found_matches = matcher(doc)
print(found_matches)
```

```python
for match_id, start, end in found_matches:
  string_id = nlp.vocab.strings[match_id]
  span = doc[start:end]
  print(match_id, string_id, start, end, span.text)
```

Rule based Matching for extracting POS

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
def extract_noun_propn(nlp_doc):
     pattern = [{'POS': 'NOUN'}, {'POS': 'PROPN'}]
     matcher.add('Noun-Propn', None, pattern)
     matches = matcher(nlp_doc)
     for match_id, start, end in matches:
         span = nlp_doc[start:end]
         return span.text
```

Constructing Tree

```python
import spacy
from nltk import Tree

en_nlp = spacy.load('en')
```

```python
doc = en_nlp("Helping others physically, by removing their physical needs, is indeed great.")
def to_nltk_tree(node):
    if node.n_lefts + node.n_rights > 0:
        return Tree(node.orth_, [to_nltk_tree(child) for child in node.children])
    else:
        return node.orth_
[to_nltk_tree(sent.root).pretty_print() for sent in doc.sents]
```
